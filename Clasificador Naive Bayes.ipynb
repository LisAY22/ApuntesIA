## Clasificador Naive Bayes

El clasificador Naïve Bayes es un algoritmo de aprendizaje automático que utiliza principios de probabilidad para clasificar datos. Se basa en el teorema de Bayes y en la hipótesis de independencia entre las variables predictoras. 
El teorema de Bayes es un método matemático que calcula la probabilidad de un evento, considerando información previa. Se basa en probabilidades condicionales.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
```

```python
# Cargar el dataset
data = sns.load_dataset("iris")
# Separación de datos en Entrenamiento.Prueba
train, test = train_test_split(data, test_size=0.4, stratify=data["species"], random_state=19)
```

```python
train
```

```python
# Separación de la etiqueta de clase y atributos
# Conjunto de entrenamiento
X_train = train[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
Y_train = train['species']
# Conjunto de pruebas
X_test = test[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
Y_test = test['species']
```

```python
# Clasificador Gaussiano Naive Bayes
mod_gnb = GaussianNB()
y_pred = mod_gnb.fit(X_train, Y_train).predict(X_test)
print("La predicción del Clasificador Gaussiano Naive Bayes en los datos de prueba es {:0.3f}".format(metrics.accuracy_score(y_pred, Y_test)))
```

```python
pred2 = ['petal_length', 'petal_width']
mod_gnb2_pred = GaussianNB()
y_pred2 = mod_gnb2_pred.fit(X_train[pred2], Y_train).predict(X_test[pred2])
print("La predicción del Clasificador Gaussiano Naive Bayes en los datos de prueba es {:0.3f}".format(metrics.accuracy_score(y_pred2, Y_test)))
```

```python
test['Prediction'] = y_pred
test['Prediction 2'] = y_pred2
test
```
