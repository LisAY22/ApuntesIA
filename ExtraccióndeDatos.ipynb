```python
import requests 
from bs4 import BeautifulSoup
```

```python
# Emitimos un request y recibimos un response
# Enlace a la página
url = 'https://es.wikipedia.org/wiki/Organizaci%C3%B3n_territorial_de_Guatemala'
```

```python
response = requests.get(url)
```

```python
print('Status code', url, ':', response.status_code)
```

```python
# Construir un objeto BeautifulSoup con el contenido html
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
tabla = soup.find('table', {'class': 'wikitable'})
```

```python
import pandas as pd
```

```python
tabla = pd.read_html(str(tabla))
tabla
```
