# Introducción a OpenCV

```python
# Importar paquetes
import numpy as np 
import imageio
import matplotlib.pyplot as plt
from skimage import color
# Biblioteca Open CV
import cv2
import common
# Para el control de dimensiones de la imagen
import pylab
```

```python
# Modificar las dimensiones de visualización
pylab.rcParams['figure.figsize'] = (6.0, 4.0)
# Cargar imagen
ruta = 'Imagenes/wall-e.jpg'
input_image = cv2.imread(ruta)
```

```python
plt.imshow(input_image)
```

```python
# OpenCV carga las imágenes en formato BGR por defecto, pero matplotlib muestra las imágenes en formato RGB. Es necesario modificarlo.
plt.imshow(cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB))
```

```python
# Inversión vertical de la imagen (0)
flippled_img = cv2.flip(input_image, 0)
plt.imshow(flippled_img);
```

```python
# Inversión horizontal de la imagen (1)
flippled_img_1 = cv2.flip(input_image, 1)
plt.imshow(flippled_img_1);
```

```python
# Imagen transpuesta
transpuesta = cv2.transpose(input_image)
plt.imshow(transpuesta);
```

```python
plt.imshow(cv2.cvtColor(cv2.flip(transpuesta, 1)  ,cv2.COLOR_BGR2RGB))
```

## Operaciones aritméticas con Open CV

```python
# Matriz de ceros del mismo tamaño que input_image rellena con 0
# Crear una matriz de h, w, 3
# Crea una matriz de ceros con las mismas dimensiones que la imagen de entrada
# blank_image es una imagen en blanco del mismo tamaño que la imagen de entrada.
blank_image = np.zeros((input_image.shape), np.uint8)
# Crear el marco fila, columna, capa
blank_image[125:255, 350:450, 1] = 50
```

```python
plt.imshow(blank_image)
```

```python
nueva = cv2.cvtColor(cv2.add(blank_image, input_image), cv2.COLOR_BGR2RGB)
plt.imshow(nueva);
```

```python
# Filtro de desnfoque 
d = 5
img_desenfocada = cv2.GaussianBlur(input_image, (2*d+1, 2*d+1), - 1)[d:-3, d:-d]
plt.imshow(cv2.cvtColor(img_desenfocada, cv2.COLOR_BGR2RGB));
```

```python
grayImage = cv2.cvtColor(input_image, cv2.COLOR_BGR2GRAY)
plt.imshow(grayImage, cmap='gray')
```

```python
sobelX = cv2.Sobel(grayImage, cv2.CV_64F, 1, 0, ksize=9)
plt.imshow(sobelX, cmap='gray');
```

```python
sobelY = cv2.Sobel(grayImage, cv2.CV_64F, 0, 1, ksize=9)
plt.imshow(sobelY, cmap='gray');
```

## Máximos y Mínimos

```python
for i in range(0,3):
    minimo, maximo, min_loc, max_loc = cv2.minMaxLoc(input_image[:,:,i])
    print('min {} esta ubicado en {} y el max {} esta ubicado en {}'.format(minimo, min_loc, maximo, max_loc))
```
