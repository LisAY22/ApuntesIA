```python
import cv2
import common
import pylab
import cv2
import pytesseract
import matplotlib.pyplot as plt
#Para windows
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"
```

```python
placa = []
image = cv2.imread('Imagenes/auto.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB));
```

```python
gray = cv2.blur(gray, (3,3))  # Desenfoque 
canny = cv2.Canny(gray, 130,200)  # Deteccin de bordes
canny = cv2.dilate(canny, None, iterations=1)   # Dilatar bordes (ms grandes) 
plt.imshow(canny, cmap='gray')
```

```python
# Deteccion de contornos con OpnenCV
# Lista de los contornos encontrados en la imagen
# El segundo valor de retorno es la jerarqua de los contornos, pero con _ se ignora
contorno, _ = cv2.findContours(canny, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
```

```python
for c in contorno:
    area = cv2.contourArea(c)
    x, y, w, h = cv2.boundingRect(c)
    epsilon = 0.09 * cv2.arcLength(c, True)   # Tolerancia, perimetro cerrado
    aprox = cv2.approxPolyDP(c, epsilon, True)  # Simplifica un contorno, circulo a cuadrado
    if len(aprox)==4 and area>=9000:
        print('area',area)
        print('x:', x)
        print('y:', y)
        print('w:', w)
        print('h:', h)
        relAspec = float(w)/h
        print('Relacion de aspecto:', relAspec)
        cv2.drawContours(image, [aprox], 0, (0,255,0), 3)  # 3 = grosor, 0 = indice del contorno
        if relAspec > 3:
            # Coordenadas del texto
            placa = gray[y:y+h, x:x+w]
            # --psm 11 es un parmetro que indica a Tesseract que espere una sola lnea de texto.
            text = pytesseract.image_to_string(placa, config='--psm 11'.strip())
            print('Placa: ', text)
            cv2.rectangle(image, (x, y), (x+w, y+h), (0,255,0), 3)
            cv2.putText(image, text, (x-20, y-20), 1, 2.2, (0,255,0), 3)  # 1 = tipo de fuente, 2.2 = tamao de fuente
```
