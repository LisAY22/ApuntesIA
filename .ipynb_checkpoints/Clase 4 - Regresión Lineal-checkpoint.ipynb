y = mx + b
<ul>
    <li>y = respuesta</li>
    <li>x = explicativa</li>
    <li>m = pendiente</li>
    <li>b = intercepto</li>
</ul>

```python
import numpy as np
import seaborn as sns
import pandas as pd
import statsmodels.api as sm
from scipy import stats
```

```python
data = sns.load_dataset('iris')
data.head(10)
```

```python
data['species'].unique()
```

```python
data['species'].value_counts()
```

```python
x = data['petal_length']
y = data['petal_width']
x
```

```python
# Agregar columna constante = 1. Para cálculo del intercepto.
x = sm.add_constant(x) 
```

```python
x
```

```python
y
```

```python
# Ajusta un modelo de regresión lineal a los datos dados y almacena el resultado del ajuste en resultado. 
resultado = sm.OLS(y, x).fit()
```

```python
resultado.summary()
```

### Correlación de Pearson (r)
Nivel de relación entre dos variables. 
<ul>
    <li>1 = Relación Directa</li>
    <li>0 = No relación</li>
    <li>-1 = Relación Negativa</li>
</ul>

### Coeficiente de Determinaión (r^2)

```python
resultado.params
```

```python
resultado.rsquared
```

```python
import math 
```

```python
math.sqrt(resultado.rsquared)
```

```python
ax = sns.regplot(data = data, x = 'petal_length', y = 'petal_width')
ax.set_xlabel('Longitud de pétalo')
ax.set_ylabel('Ancho de pétalo')
ax.set_title('Relación entre la Longitud y el Ancho del pétalo')
ax.plot();
```

```python
ax2 = sns.jointplot(data = data, x = 'petal_length', y = 'petal_width', kind = 'reg', truncate = False, color = 'm', height = 7)
```
